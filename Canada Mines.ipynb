```python
def popup_html(row):
    i = row
    operation = mines['Operation '].iloc[i]
    operator = mines['Operator / Owners'].iloc[i] 
    commodity_group = mines['Commodity Group'].iloc[i]
    commodity = mines['Commodity'].iloc[i] 
    
    
    left_col_color = "#526f8a"
    right_col_color = "#A7E21D"
    
    html = """<!DOCTYPE html>
<html>
<head>
<h4 style="margin-bottom:10"; width="200px">{}</h4>""".format(operation) + """
</head>
    <table style="height: 126px; width: 350px;">
<tbody>
<tr>
<td style="background-color: """+ left_col_color +""";"><span style="color: #ffffff;">Operation</span></td>
<td style="width: 150px;background-color: """+ right_col_color +""";">{}</td>""".format(operation ) + """
</tr>
<tr>
<td style="background-color: """+ left_col_color +""";"><span style="color: #ffffff;">Owner/Operator</span></td>
<td style="width: 150px;background-color: """+ right_col_color +""";">{}</td>""".format(operator) + """
</tr>
<tr>
<td style="background-color: """+ left_col_color +""";"><span style="color: #ffffff;">Category</span></td>
<td style="width: 150px;background-color: """+ right_col_color +""";">{}</td>""".format(commodity_group) + """
</tr>
<tr>
<td style="background-color: """+ left_col_color +""";"><span style="color: #ffffff;">Commodity</span></td>
<td style="width: 150px;background-color: """+ right_col_color +""";">{}</td>""".format(commodity) + """
</tr>
</tbody>
</table>
</html>
"""
    return html
```

```python
location = mines['lat'].mean(), mines['long'].mean()
map = folium.Map(location=location, tiles='OpenStreetMap', zoom_start=3)

for i in range(0, len(mines)):
    html = popup_html(i)
    iframe = branca.element.IFrame(html=html, width=510, height=200)
    popup = folium.Popup(folium.Html(html, script=True), max_width=500)
    folium.Marker([mines['lat'].iloc[i],mines['long'].iloc[i]],
                  popup=popup,icon=folium.Icon(color="green", icon='tree', prefix='fa')).add_to(map)
    
map
```
